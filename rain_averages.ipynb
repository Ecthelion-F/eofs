##### 结论
+ ~~产生正峰值请用降水偏多替代，产生负峰值用降水偏少代替~~
+ ~~全国平均不明显，整体上存在一个下降趋势，结合小波分析结果可能是更大时间尺度震荡的一部分~~
+ 东北地区在拉尼娜后的厄尔尼诺结束年会产生正峰值，厄尔尼诺连续出现时结束年会产生负峰，前期没有异常的厄尔尼诺没有马
+ 华北之大已安放不下一张平静的课桌了
+ 2015厄尔尼诺没有马
+ 华中有高达85%的概率在厄尔尼诺的结束年出正峰值，不区分中东部
+ 华南地区东部型厄尔尼诺结束年有负峰值
+ 长江以南，东部厄尔尼诺，看终值后面那年，显著正相关
+ 厄尔尼诺结束年次年，东北降水减少
